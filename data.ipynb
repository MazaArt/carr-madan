```python
import yfinance as yf
import pandas as pd
import scipy
```

```python
stocks = {
    "Apple" : "AAPL",
    "Microsoft" : "MSFT",
    "Nvidia" : "NVDA",
    "Palantir" : "PLTR",
    "Google" : "GOOG"
}
```

```python
tkr = "^SPX"
stock_data = yf.Ticker(tkr)
print(stock_data.option_chain(stock_data.options[0])._fields)
# fields of option - calls, puts, underlying
```

```python
# properties of puts and calls
props = ['contractSymbol', 'lastTradeDate', 'strike', 'lastPrice', 'bid', 'ask',
       'change', 'percentChange', 'volume', 'openInterest',
       'impliedVolatility', 'inTheMoney', 'contractSize', 'currency']

print(stock_data.option_chain(stock_data.options[0]).calls.columns)

print(stock_data.option_chain(stock_data.options[0]).puts.columns)

underlying_props = list(stock_data.option_chain(stock_data.options[0]).underlying.keys())

# print underlying properties
for i in range(len(underlying_props)):
    print(f"{underlying_props[i]}", end = ", ")
    if i > 0 and i % 5 == 0:
       print()
```

```python
exp_dat_map = dict() # map date to idx of stock_data.options 

for idx, exp in enumerate(stock_data.options): # this consists of a tuple of dates that is only iterable
    exp_dat_map[exp] = idx

# Note: we can only pull future pricing data 
future_date = "2026-04-10"
exp_arg = exp_dat_map[future_date]

# why are there more put options than call options 
print(stock_data.option_chain(stock_data.options[exp_arg]).calls)
print(stock_data.option_chain(stock_data.options[exp_arg]).puts)
```

```python
# ex tkr: 'NVDA'
# ex date: '2026-03-26' (YYYY-MM-DD)
def get_call_data(tkr, date):
    stock_data = yf.Ticker(tkr)
    return stock_data.option_chain(stock_data.options[exp_dat_map[date]]).calls

def get_put_data(tkr, date):
    stock_data = yf.Ticker(tkr)
    return stock_data.option_chain(stock_data.options[exp_dat_map[date]]).puts

# print(get_call_data('NVDA', '2026-04-01'))
# print(get_put_data('NVDA', '2026-04-01'))
print(get_put_data('^SPX', future_date))

```
